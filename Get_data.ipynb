```python
import os
import shutil

def download_from_AFDB(fasta_dir,pid,working_dir):
    if not os.path.exists(working_dir):
        os.mkdir(working_dir)
    print(f"Protein id in AlphaFold DB is: {pid}")
    alphafold_ID = f'AF-{pid}-F1'
    database_version = 'v4'
    model_url = f'https://alphafold.ebi.ac.uk/files/{alphafold_ID}-model_{database_version}.pdb'
    if not os.path.exists('fasta_prot_need_runAF2'):
        os.mkdir('fasta_prot_need_runAF2')
    f = open('fasta_prot_need_runAF2/pids.txt','w')
    
    if not os.path.exists(f"{working_dir}/{alphafold_ID}.pdb"):
        os.system(f'curl {model_url} -f -o {working_dir}/{alphafold_ID}.pdb')
        
    if not os.path.exists(f"{working_dir}/{alphafold_ID}.pdb"):
        print(f'I did not succeed download PDB files, please check pid: {pid} :(')
        print(f'{fasta_dir}/{pid}.fasta')
        # os.system(f'cp {fasta_dir}/{pid}.fasta fasta_prot_need_runAF2')
        shutil.copy(f'{fasta_dir}/{pid}.fasta',f'fasta_prot_need_runAF2/')
        f.write(f'{pid} ')
    else:
        print('I downloaded PDB files :)')
        
    f.close()
        
    output_pdb = f'{working_dir}/{alphafold_ID}.pdb'
    return output_pdb
```

```python
import os

def read_fasta_files(directory_path):
    """
    Reads all FASTA files in the given directory and returns a dictionary where
    the keys are file names and the values are the protein sequences contained in those files.

    Args:
    directory_path (str): The path to the directory containing the FASTA files.

    Returns:
    dict: A dictionary with file names as keys and protein sequences as values.
    """
    pdb_seq_dict = {}
    
    for filename in os.listdir(directory_path):
        if filename.endswith(".fasta") or filename.endswith(".fa"):
            file_path = os.path.join(directory_path, filename)
            with open(file_path, 'r') as file:
                lines = file.readlines()
                protein_sequence = ''.join([line.strip() for line in lines if not line.startswith(">")])
                pid = filename.split('.')[0]
                key = f"AF-{pid}-F1.pdb"
                pdb_seq_dict[key] = protein_sequence

    return pdb_seq_dict
```

# Get csv out of PDB file

```python
def write_dict_to_csv(pdb_seq_dict, output_file):
    """
    Writes a dictionary to a CSV file. The keys are assumed to be the names of the FASTA files,
    and the values are the protein sequences.

    Args:
    fasta_dict (dict): The dictionary containing file names as keys and protein sequences as values.
    output_file (str): The path to the output CSV file.
    """
    with open(output_file, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["PDB_File_Name", "Protein_Sequence"])  # Writing the header
        for file_name, sequence in pdb_seq_dict.items():
            writer.writerow([file_name, sequence])
    print(f"Sequences saved to {output_file} :) ")
```
